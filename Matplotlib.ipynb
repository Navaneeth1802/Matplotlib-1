```python
import pandas as pd
df=pd.read_csv('Downloads/data.csv')
df.head(5)
```

```python
Age=df['Age']
Devs=df['All_Devs']
Python_dev=df['Python']
Java_Dev=df['JavaScript']
```

```python
plt.plot(Age,Devs,linestyle='--',label='ALLDEVS')
plt.plot(Age,Python_dev,linestyle='--',label='Python_dev')
plt.plot(Age,Java_Dev,linestyle='-',label='JavaDev')
plt.legend()
median_salary=60000
plt.fill_between(Age,Python_dev,median_salary,where=(Python_dev>median_salary),interpolate=True,alpha=0.35)
plt.fill_between(Age,Python_dev,median_salary,where=(Python_dev<median_salary),interpolate=True,alpha=0.25)
```

```python
plt.plot(Age,Python_dev,label='Python Developers')
plt.plot(Age,Java_Dev,label='Java  Developers')
plt.fill_between(Age,Python_dev,Java_Dev,where=(Java_Dev>Python_dev),interpolate=True,alpha=0.5)
plt.legend()
```
